```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your weather data from the CSV file
data = pd.read_csv('indiawheather.csv')

# Filter data for Rajasthan
rajasthan_data = data[data['state'] == 'Rajasthan']

# Calculate average temperature by city in Rajasthan
rajasthan_avg_temp_by_city = rajasthan_data.groupby('city_name')['temp_max'].mean().reset_index()

# Sort the data to highlight top 5 cities
rajasthan_avg_temp_by_city_sorted = rajasthan_avg_temp_by_city.sort_values(by='temp_max', ascending=False)

# Set up the figure and size
plt.figure(figsize=(10, 6))

# Define colors based on conditions
colors = ['skyblue' if city not in rajasthan_avg_temp_by_city_sorted.iloc[:5]['city_name'].values else 'violet' for city in rajasthan_avg_temp_by_city['city_name']]

# Plotting the bar chart using Seaborn with customized colors
sns.barplot(x='city_name', y='temp_max', data=rajasthan_avg_temp_by_city, palette=colors)

# Adding labels and title
plt.title('Average Tempreture by City in Rajasthan')
plt.xlabel('City')
plt.ylabel('Average Tempreture')
plt.xticks(rotation=45)

# Adding legend at lower right position
legend_labels = ['Other Cities (Tempreture)', 'Top 5 Cities (MAX Tempreture)']
legend_handles = [plt.Rectangle((0,0),1,1, color=color) for color in ['skyblue', 'violet']]
plt.legend(legend_handles, legend_labels, loc='lower right')

# Ensure labels and title fit within the figure area
plt.tight_layout()

# Show the plot
plt.show()
```
