```python
import matplotlib.pyplot as plt
import pandas as pd

# Read data from CSV file
data = pd.read_csv('indiawheather.csv')

# Filter data for Kolkata
kolkata_data = data[data['city_name'] == 'Kolkata']

# Convert dt_txt column to datetime format
kolkata_data['dt_txt'] = pd.to_datetime(kolkata_data['dt_txt'], format='%d-%m-%Y %H:%M')

# Group by date and calculate mean wind speed and direction
grouped_data = kolkata_data.groupby(kolkata_data['dt_txt'].dt.date).agg({
    'wind_speed': 'mean',
    'wind_degree': 'mean'
}).reset_index()

# Extract data from grouped DataFrame
dates = grouped_data['dt_txt']
wind_speeds = grouped_data['wind_speed']
wind_directions = grouped_data['wind_degree']

# Create a figure and axis for the plot
fig, ax1 = plt.subplots(figsize=(14, 8))

# Plot wind speed on the first y-axis (left)
ax1.plot(dates, wind_speeds, marker='o', linestyle='-', color='blue', label='Average Wind Speed (m/s)')
ax1.set_xlabel('Date')
ax1.set_ylabel('Wind Speed (m/s)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Customize y-axis ticks for wind speed
ax1.set_yticks(range(0, int(max(wind_speeds)) + 10,1 ))

# Create a second y-axis (right) for wind direction
ax2 = ax1.twinx()
ax2.plot(dates, wind_directions, marker='s', linestyle='-', color='green', label='Average Wind Direction (degrees)')
ax2.set_ylabel('Wind Direction (degrees)', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Customize y-axis ticks for wind direction
ax2.set_yticks(range(0, int(max(wind_directions)) + 20, 10))

# Set labels and title
plt.title('Average Wind Speed and Wind Direction in Kolkata')
fig.tight_layout()
plt.grid(True)

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.show()
```
